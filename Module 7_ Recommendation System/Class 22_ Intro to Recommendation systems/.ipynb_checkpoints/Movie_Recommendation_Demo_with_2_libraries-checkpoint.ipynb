```python
import pandas as pd
from surprise import Dataset, Reader, SVD, accuracy
from surprise.model_selection import train_test_split

# Load the movielens-100k dataset (download it if needed),
data = Dataset.load_builtin('ml-100k')

# The Reader object helps in parsing the file or dataframe containing ratings.
reader = Reader(line_format='user item rating timestamp', sep='\t')

# Split the dataset into the train and test set.
trainset, testset = train_test_split(data, test_size=0.25)

# Use the famous SVD algorithm.
algo = SVD()

# Train the algorithm on the trainset.
algo.fit(trainset)

# Predict ratings for the testset.
predictions = algo.test(testset)

# Compute and print the Mean Squared Error
accuracy.rmse(predictions)

# To make predictions for a specific user and item.
uid = str(196)  # predict the rating for user with uid '196'
iid = str(302)  # and movie with iid '302'
pred = algo.predict(uid, iid, r_ui=4, verbose=True)
```

```python
import os
import tempfile

import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_recommenders as tfrs
```

```python
class MovieLensModel(tfrs.Model):

    def __init__(self, user_model, movie_model):
        super().__init__()
        self.movie_model: tf.keras.Model = movie_model
        self.user_model: tf.keras.Model = user_model
        self.task: tf.keras.layers.Layer = tfrs.tasks.Retrieval()

    def compute_loss(self, features, training=False):
        # We pick out the user features and pass them into the user model.
        user_embeddings = self.user_model(features["user_id"])
        # And pick out the movie features and pass them into the movie model,
        # getting embeddings back.
        positive_movie_embeddings = self.movie_model(features["movie_title"])

        # The task computes the loss and the metrics.
        return self.task(user_embeddings, positive_movie_embeddings)
```

```python
ratings = tfds.load('movielens/100k-ratings', split="train")
movies = tfds.load('movielens/100k-movies', split="train")
```

```python
# Select the necessary attributes
ratings = ratings.map(lambda x: {
    "movie_title": x["movie_title"],
    "user_id": x["user_id"]
})
movies = movies.map(lambda x: x["movie_title"])
```

```python
# Prepare the data for training
user_ids_vocabulary = tf.keras.layers.StringLookup()
user_ids_vocabulary.adapt(ratings.map(lambda x: x["user_id"]))

movie_titles_vocabulary = tf.keras.layers.StringLookup()
movie_titles_vocabulary.adapt(movies)
```

```python
# Shuffle, batch, and cache the dataset.
cached_train = ratings.shuffle(100_000).batch(8192).cache()
cached_test = ratings.batch(4096).cache()
```

```python
# Define user and movie models
user_model = tf.keras.Sequential([
    user_ids_vocabulary,
    tf.keras.layers.Embedding(user_ids_vocabulary.vocabulary_size(), 64)
])

movie_model = tf.keras.Sequential([
    movie_titles_vocabulary,
    tf.keras.layers.Embedding(movie_titles_vocabulary.vocabulary_size(), 64)
])
```

```python
# Instantiate the model
model = MovieLensModel(user_model, movie_model)

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adagrad(learning_rate=0.1))
```

```python
# Train the model
model.fit(cached_train, epochs=3)
```

```python
# Evaluate the model
model.evaluate(cached_test, return_dict=True)
```
