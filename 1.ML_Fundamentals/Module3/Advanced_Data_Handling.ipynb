# Data Visualization Techniques


### Histograms and Frequency Distribution

- Shows distribution of a set of continous data

```python
import matplotlib.pyplot as plt

data = [23, 25, 26, 26, 27, 29, 30, 31, 32, 35, 36, 38, 40, 42]
plt.hist(data, bins=5, edgecolor="k")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Bar and Column charts

- Compare categories of data

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C']
sales = [100, 150, 80]
plt.bar(products, sales, color=['blue', 'green', 'red'])
plt.title("Sales of Different Products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

### Line and Area Charts

- Display trends over time

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
visitors = [1000, 1400, 1200, 1600, 1800]
plt.plot(months, visitors, marker='o')
plt.fill_between(months, visitors, color='skyblue', alpha=0.4)  # for area under the line
plt.title("Monthly Website Visitors")
plt.xlabel("Month")
plt.ylabel("Visitors")
plt.show()
```

### Scatter Plots

- Used to display relationship between two quantitative variables

```python
import matplotlib.pyplot as plt

ages = [25, 30, 35, 40, 45]
income = [40000, 50000, 60000, 70000, 80000]
plt.scatter(ages, income, color='purple')
plt.title("Relationship between Age and Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

### Box and Whisker Plots

- Shows distribution of data based on 5 summary statistics

```python
import seaborn as sns

data = [20, 22, 23, 24, 25, 29, 30, 32, 35, 38, 40, 42]
sns.boxplot(data=data)
plt.title("Boxplot of Data Distribution")
plt.show()
```

### Heatmaps
- Shows correlation between features in datasets

```python
import seaborn as sns
import numpy as np

# Sample data: correlation matrix
data = np.array([
    [1, 0.8, 0.5],
    [0.8, 1, 0.3],
    [0.5, 0.3, 1]
])
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Violin plots
- Used to show distribution of data across different categories

```python
import seaborn as sns

classes = ['Class A', 'Class B', 'Class C']
scores = [85, 90, 88, 78, 92, 85, 96, 87, 80, 83, 91, 88]
class_category = ['Class A'] * 4 + ['Class B'] * 4 + ['Class C'] * 4

sns.violinplot(x=class_category, y=scores)
plt.title("Distribution of Exam Scores across Classes")
plt.show()
```

# Interactive Visuals


```python
import plotly.express as px

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()
```
